Incorporation de divers dataset en un seul pour train 


```python
import pandas as pd
import numpy as np
from tqdm import tqdm
from matplotlib import pyplot as plt

file = "Sarcasm_Headlines_Dataset.csv"
file2 = "train.En.csv"
file3 = "sarcasm2.csv"


df = pd.read_csv(file, index_col=0)
df2 = pd.read_csv(file2, index_col=0)
df3 = pd.read_csv(file3, index_col=0)

```

```python
# Remove column name 'rephrase','sarcasm','irony','satire','understatement','overstatement','rhetorical_question'
df2=df2.drop(['rephrase','sarcasm','irony','satire','understatement','overstatement','rhetorical_question'], axis=1)

df2 #Affiche les 5 premières lignes de notre dataframe
```

```python
# Remove column name 'article_link','Unnamed: 3','Unnamed: 4'
df=df.drop(df.columns[[2,3]], axis=1)
df.rename(columns = {'text':'tweet'}, inplace = True)

df = df[(df['sarcastic'] == '0') | (df['sarcastic'] == '1')]
df
```

```python
df3['sarcastic'] = np.where(df3['class']== 'sarcasm', 1, 0)
df3=df3.drop('class', axis=1)
df3['tweet']=df3.index
pd.unique(df3['sarcastic'])
```

```python
final_df=pd.concat([df2, df,df3], axis=0)
final_df=final_df.reset_index()
final_df=final_df.drop('index', axis=1)
final_df['sarcastic'] = final_df['sarcastic'].astype(int)
```
